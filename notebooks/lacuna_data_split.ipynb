```python
import os
import random
import shutil
```

```python
# data folders
dir_healthy = "../data/lacuna_combined/healthy"
dir_farmyworm = "../data/lacuna_combined/fallarmyworm"
```

## Helper function

```python
def sample_data_files(src, dst, no_samples):
    for file in random.sample(os.listdir(src), no_samples):
        #check if file is jpg image
        if ".jpg" in file:
            file_path = os.path.join(src, file)
            #copy file
            print(f"Copying file {file_path} to {dst}")
            shutil.move(file_path, dst)
```

```python
train_h = "../data/data1/train/healthy"
train_fa= "../data/data1/train/fallarmyworm"
test_h = "../data/data1/test/healthy"
test_fa = "../data/data1/test/fallarmyworm"
val_h = "../data/data1/validation/healthy"
val_fa = "../data/data1/validation/fallarmyworm"
```

##  Train set

```python
#healthy
sample_data_files(dir_healthy, train_h, 800)
#fallarmyworm
sample_data_files(dir_farmyworm, train_fa, 800)
```

## Validation set

```python
#healthy
sample_data_files(dir_healthy, val_h, 100)
#fallarmyworm
sample_data_files(dir_farmyworm, val_fa, 100)
```

## Test set

```python
#healthy
sample_data_files(dir_healthy, test_h, 349)
#fallarmyworm
sample_data_files(dir_farmyworm, test_fa, 349)
```
